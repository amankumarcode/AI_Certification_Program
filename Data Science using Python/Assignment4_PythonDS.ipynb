```python
pd_data1 = pd_data1.iloc[:,0:3] #removing timestamp
pd_data1
```

```python
pd_data1 = pd_data1.pivot(index ='userId', columns ='movieId') #converting to a 2D array
pd_data1
```

```python
# QUESTION 1
def norm_vec(a):                  #Normalization function
    p=a/np.sqrt(np.sum(a**2))
    return p
u1=pd.to_numeric(pd_data1.iloc[0,0:],errors='coerce') #row1 i.e user1
u2=pd.to_numeric(pd_data1.iloc[1,0:],errors='coerce') #row2 i.e user2

norm_u1=norm_vec(u1)
norm_u2=norm_vec(u2)

def dot_prod(a,b):                          #dot product function of two vectors, replacing nan by mean of the vector.
    d=np.where(np.isnan(a),a.mean(axis = 0),a).dot(np.where(np.isnan(b),b.mean(axis = 0),b))
    return d

dot_p=dot_prod(norm_u1,norm_u2) #dot product of two rows
dot_p
```

```python
# QUESTION 2
c1=pd.to_numeric(pd_data1.iloc[:,0],errors='coerce') #column1 i.e movie1
c2=pd.to_numeric(pd_data1.iloc[:,1],errors='coerce') #column2 i.e movie2
norm_c1=norm_vec(c1)
norm_c2=norm_vec(c2)
dot_p=dot_prod(norm_c1,norm_c2) # dot product of two columns
dot_p
```

```python
#QUESTION 3 - PEARSON'S SIMILARITY ROWS
def norm_vec_pcc(e):
    m=np.mean(e)
    d=e-m
    return d
v1=pd.to_numeric(pd_data1.iloc[0,0:],errors='coerce')
norm_v1=norm_vec_pcc(v1)

v2=pd.to_numeric(pd_data1.iloc[1,0:],errors='coerce')
norm_v2=norm_vec_pcc(v2)

s_norm_v1v2=np.sum(norm_v1*norm_v2)
s_norm_v1v1=np.sum(norm_v1**2)
s_norm_v2v2=np.sum(norm_v2**2)
pcc_v1v2=s_norm_v1v2/np.sqrt(s_norm_v1v1*s_norm_v2v2)
pcc_v1v2
```

```python
#QUESTION 4 - PEARSON'S SIMILARITY COLUMNS
c3=pd.to_numeric(pd_data1.iloc[:,0],errors='coerce')
norm_c3=norm_vec_pcc(c3)
c4=pd.to_numeric(pd_data1.iloc[:,1],errors='coerce')
norm_c4=norm_vec_pcc(c4)
s_norm_c3c4=np.sum(norm_c3*norm_c2)
s_norm_c3c3=np.sum(norm_c3**2)
s_norm_c4c4=np.sum(norm_c4**2)
pcc_c3c4=s_norm_c3c4/np.sqrt(s_norm_c3c3*s_norm_c4c4)
pcc_c3c4
```

```python
#QUESTION 5 - ADJUSTED COSINE SIMILARITY BETWEEN USERS
pd_data1_numeric=pd_data1.apply(pd.to_numeric,errors='coerce')
def col_avg(df):
    avg=df.iloc[:,0:].mean(axis=0)
    return avg

col_mean=col_avg(pd_data1_numeric)
col_mean
def norm_vec_col(p,col_mean):
    b=p-col_mean
    p=b/np.sqrt(np.sum(b**2))
    return p

norm_v3=norm_vec_col(v1,col_mean)
norm_v4=norm_vec_col(v2,col_mean)
dot_p=dot_prod(norm_v3,norm_v4)
dot_p
```

```python
#QUESTION 6 - ADJUSTED COSINE SIMILARITY BETWEEN MOVIES
pd_data1_numeric=pd_data1.apply(pd.to_numeric,errors='coerce')
def row_avg(df):
    avg=df.iloc[:,0:].mean(axis=1)
    return avg

row_mean=row_avg(pd_data1_numeric)
row_mean
def norm_vec_row(q,row_mean):
    f=q-row_mean
    x=f/np.sqrt(np.sum(f**2))
    return x

c5=pd_data1_numeric.iloc[:,0]
c6=pd_data1_numeric.iloc[:,1]
norm_c5=norm_vec_row(c5,row_mean)
norm_c6=norm_vec_row(c6,row_mean)
dot_p=dot_prod(norm_c5,norm_c6)
dot_p
```
