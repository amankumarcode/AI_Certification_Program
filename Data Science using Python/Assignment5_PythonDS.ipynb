```python
#Question1 Part(C)
liver2 = liver.drop('dose',axis= 1)
liver2
```

```python
import pandas as pd
import numpy as np
def min_max_norm(v):   
    l = list()
    for i,val in enumerate(v):
        j = val- min(liver2.iloc[:,i])
        k =  max(liver2.iloc[:,i])- min(liver2.iloc[:,i])
        m = j/k
        l.append(m)
    l = np.array(l)
    return l

u0 = liver2.iloc[0]
u1 = liver2.iloc[1]


row1_norm = min_max_norm(u0)
row2_norm = min_max_norm(u1)

dist = np.linalg.norm(row1_norm-row2_norm)
dist
```

```python
rna = pd.read_csv(r"C:\Users\Mahe\Downloads\data_mRNA_subset.txt",delimiter = '\t')
rna
```

```python
rna1 = rna.iloc[:,1:]     #REMOVING ID_REF COLUMN
rna1                          
```

```python
# Question2 Part(A)
sum = sum(rna1.var(axis=1) <= 0.2)
if  sum >= 1 :
    rna2 = rna1[rna1.var(axis=1) > 0.2]
    print(rna2)
else:
    print("no such probe-id is found!!!")
```

```python
#Question2 Part(B) - Spearman Rank
df1 = rna1.rank(axis = 1)  
df2 = df1.iloc[0:2,:]    #Selecting first two probe IDs
df2
```

```python
r1 = df2.iloc[0,:]
r2 = df2.iloc[1,:]
d = r1-r2
d = d**2
d = d.sum()
n = len(rna1.columns)
z = (6*d)/(n*((n**2)-1))
p = 1-z
p           #Spearman Coefficient
```
