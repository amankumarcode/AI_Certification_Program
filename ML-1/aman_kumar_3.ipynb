```python
import pandas as pd
data = pd.read_csv(r'C:\Users\Mahe\Downloads\data3.csv')
data
```

```python
#RENAMING THE COLUMNS AS F1 - FEATURE 1 AND SO ON AND LAST COLUMN IS NAMED AS CLASS
data.columns = ['F1','F2','F3','F4','F5','F6','F7','F8','Class']
data
```

```python
# no of classes
num_classes = len(data['Class'].unique())
num_classes
```

```python
grouped_data = data.groupby('Class')
```

```python
class_summary_1 = [] # CONDITIONAL PROBABILITIES OF EACH FEATURE FOR CLASS 0
class_summary_2 = [] # CONDITIONAL PROBABILITIES OF EACH FEATURE FOR CLASS 1
for feature in data.iloc[:,:-1]:
    Prob1= ((grouped_data.get_group(0)[feature] == 0).sum()+1)/((data['Class'] == 0).sum()+2) #Laplace smoothing with alpha=1
    Prob2= ((grouped_data.get_group(0)[feature] == 1).sum()+1)/((data['Class'] == 0).sum()+2) #Laplace smoothing with alpha=1
    Prob3= ((grouped_data.get_group(1)[feature] == 0).sum()+1)/((data['Class'] == 1).sum()+2) #Laplace smoothing with alpha=1
    Prob4= ((grouped_data.get_group(1)[feature] == 1).sum()+1)/((data['Class'] == 1).sum()+2) #Laplace smoothing with alpha=1
    class_summary_1.append([Prob1,Prob2])
    class_summary_2.append([Prob3,Prob4])
```

```python
# POSTERIOR PROBABILITY FUNCTION IF CLASS IS ZERO
import numpy as np
def posterior_probability_class_zero(sample,summary1):
    prob = 1
    for i, feature in enumerate(sample):
        prob*= summary1[i][feature]
    prob*= priori_zero
    return prob
```

```python
# POSTERIOR PROBABILITY FUNCTION IF CLASS IS ONE
def posterior_probability_class_one(sample,summary2):
    prob = 1
    for i, feature in enumerate(sample):
        prob*= summary2[i][feature]
    prob*= priori_one
    return prob
```

```python
def predict(probabilities): # return class with max probability
    return np.argmax(probabilities)
```

```python
# GIVES THE PRIOR PROBABILITIES OF BOTH CLASSES FROM TRAINING SET
def prior(num):
    prior = len(grouped_data.get_group(num))/(len(data['Class']))
    return prior
```

```python
#TAKING OUT PRIORS OF BOTH THE CLASSES
priori_zero = prior(0)
priori_one = prior(1)
```

```python
test_data = pd.read_csv(r'C:\Users\Mahe\Downloads\test3.csv')
test_data
```

```python
test_data.columns = ['F1','F2','F3','F4','F5','F6','F7','F8']
test_data
```

```python
predicted_classes = []
num_test_samples = len(test_data)
for i in range(num_test_samples):
    probabilities = []
    sample = test_data.iloc[i] # takes a single test instance
    posterior_zero = posterior_probability_class_zero(sample,class_summary_1) #gets the probability of test instance for class zero
    probabilities.append(posterior_zero)
    posterior_one = posterior_probability_class_one(sample,class_summary_2) #gets the probability of test instance for class one
    probabilities.append(posterior_one)
    prediction = predict(probabilities) # gives out the class which has maximum probability for given test instance
    predicted_classes.append(prediction) 
```
